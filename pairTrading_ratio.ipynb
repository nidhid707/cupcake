```python
trade_short = []
trade_long  = []
Train_collect = []
Test_collect = []

!pip install yfinance
import yfinance as yf

for pair in PAIR_LIST.index:
    #exists1 = os.path.isfile(pathDaily + PAIR_LIST["stock 1"][pair]+'.csv')
    #exists2 = os.path.isfile(pathDaily + PAIR_LIST["stock 2"][pair]+'.csv')
    
    #if exists1:
     #   if exists2:
    # Load configuration file values
      #      stock1 = pd.read_csv(pathDaily + PAIR_LIST["stock 1"][pair]+'.csv')
       
       #     stock1=stock1.reindex(index=stock1.index[::-1])
            nsedata = yf.Ticker(PAIR_LIST["stock 1"][pair]+".NS")

            stock1 = nsedata.history(period="2y")
            
            stock1["Time"] = stock1.index
            stock1["Time"] = pd.to_datetime(stock1.Time, format='%Y-%m-%d %H:%M:%S')
            stock1 = stock1[['Time', 'Open', 'Close']]
            stock1.columns = ['Time', 'Stock1_Open', 'Stock1_Close']
            stock1["stock1"]=PAIR_LIST["stock 1"][pair]
            #stock1=stock1.reindex(index=stock1.index[::-1])
            print(stock1.tail())


            #stock2 = pd.read_csv(pathDaily + PAIR_LIST["stock 2"][pair]+'.csv')
            #stock2=stock2.reindex(index=stock2.index[::-1])
            nsedata = yf.Ticker(PAIR_LIST["stock 2"][pair]+".NS")

            stock2 = nsedata.history(period="2y")
            
            stock2["Time"] = stock2.index
            stock2["Time"] = pd.to_datetime(stock2.Time, format='%Y-%m-%d %H:%M:%S')
            stock2 = stock2[['Time', 'Open', 'Close']]
            stock2.columns = ['Time', 'Stock2_Open', 'Stock2_Close']
            stock2["stock2"]=PAIR_LIST["stock 2"][pair]
            print(stock2.tail())

            merged = pd.merge(stock1, stock2, on='Time')
            merged=merged.dropna()
            merged=merged[::-1].reset_index()
            
            mask = (merged['Time'] > '2019-01-01') & (merged['Time'] < '2019-12-31')
            Train = merged.loc[mask]
            print(Train.tail())
            Train.loc[:,'Ratio'] = Train.Stock1_Close / Train.Stock2_Close
            pair_mean = statistics.mean(Train['Ratio'])
            pair_std = np.std(Train['Ratio'], ddof=0)
            plusdev = pair_mean + pair_std
            minusdev = pair_mean - pair_std
            corr=scipy.stats.pearsonr(Train['Stock1_Close'], Train['Stock2_Close'])[0]
            score, pvalue, _ = coint(Train["Stock1_Close"],Train["Stock2_Close"])
                
            if (corr>0.85):
                #if (pvalue<0.05):
                
                    print(PAIR_LIST["stock 1"][pair],PAIR_LIST["stock 2"][pair],"correlation:",corr)
                    print("Cointegration pvalue: ",pvalue)
            
            
                    Train.loc[:,"density curve"] = scipy.stats.norm.cdf(Train['Ratio'], loc = pair_mean, scale = pair_std)
    
                    mask2 = (merged['Time'] > '2020-01-01') & (merged['Time'] < '2020-12-31')
                    Test = merged.loc[mask2]
            
                    Test.loc[:,'Correlation']=corr
                    Test.loc[:,'Cointegrtion(pvalue)']=pvalue 
                
                    Test.loc[:,'Ratio'] = Test.Stock1_Close / Test.Stock2_Close
                    Test.loc[:,"density curve"] = scipy.stats.norm.cdf(Test['Ratio'], loc = pair_mean, scale = pair_std)
                
                    Test.loc[:,"Long_Trade"] = np.where(Test["density curve"] < 0.02,"Entry",np.where((Test['Ratio'] <= (pair_mean-0.00001)) & (Test['Ratio'] >= (minusdev-0.00001)),"Exit","No trade"))
                    Test.loc[:,"long Buy stock1"] = np.where((Test["Long_Trade"].shift(-1) == "Entry")&((Test["Long_Trade"].shift(-2) == "Exit")|(Test["Long_Trade"].shift(-2) == "No trade")),Test["Stock1_Open"],0)
                    Test.loc[:,"long Sell stock2"] = np.where((Test["Long_Trade"].shift(-1) == "Entry")&((Test["Long_Trade"].shift(-2)=="Exit")|(Test["Long_Trade"].shift(-2) == "No trade")),Test["Stock2_Open"],0.00)
                    Test.loc[:,"long stock1 qty"] = np.where(Test["long Buy stock1"] != 0,100000 / Test["long Buy stock1"],0)
                    Test.loc[:,"long stock2 qty"] = np.where(Test["long Sell stock2"] != 0,100000 / Test["long Sell stock2"],0)
                    Test.loc[:,"long Buy stock2"] = np.where(((Test["Long_Trade"] == "Exit") & ((Test["Long_Trade"].shift(-1) == "Entry")|(Test["Long_Trade"].shift(-1) == "No trade"))),Test["Stock2_Close"],0) 
                    Test.loc[:,"long Sell stock1"] = np.where(((Test["Long_Trade"] == "Exit") & ((Test["Long_Trade"].shift(-1) == "Entry")|(Test["Long_Trade"].shift(-1) == "No trade"))),Test["Stock1_Close"],0)
    
    
                    Test.loc[:,"Short_Trade"] = np.where(Test["density curve"] > 0.97,"Entry",np.where((Test['Ratio'] >= (pair_mean+0.00001)) & (Test['Ratio'] <= (plusdev+0.00001)),"Exit","No trade"))
                    Test.loc[:,"short Sell stock1"] = np.where((Test["Short_Trade"].shift(-1) == "Entry")&((Test["Short_Trade"].shift(-2) == "Exit")|(Test["Short_Trade"].shift(-2) == "No trade")),Test["Stock1_Open"],0.00)
                    Test.loc[:,"short Buy stock2"] = np.where((Test["Short_Trade"].shift(-1) == "Entry")&((Test["Short_Trade"].shift(-2) == "Exit")|(Test["Short_Trade"].shift(-2) == "No trade")),Test["Stock2_Open"],0)
                    Test.loc[:,"short stock1 qty"] = np.where(Test["short Sell stock1"] != 0 ,100000/Test["short Sell stock1"],0)
                    Test.loc[:,"short stock2 qty"] = np.where(Test["short Buy stock2"] != 0 ,100000/Test["short Buy stock2"],0)
                    Test.loc[:,"short Buy stock1"] = np.where(((Test["Short_Trade"] == "Exit")&((Test["Short_Trade"].shift(-1) == "Entry")|(Test["Short_Trade"].shift(-1) == "No trade"))),Test["Stock1_Close"],0) 
                    Test.loc[:,"short Sell stock2"] = np.where(((Test["Short_Trade"] == "Exit")&((Test["Short_Trade"].shift(-1) == "Entry")|(Test["Short_Trade"].shift(-1) == "No trade"))),Test["Stock2_Close"],0)
    
                    df1 = Test[((Test["Short_Trade"].shift(-1) == "Entry") & ((Test["Short_Trade"].shift(-2) == "Exit")|(Test["Short_Trade"].shift(-2) == "No trade")))|(((Test["Short_Trade"].shift(-2) == "Exit")) & ((Test["Short_Trade"].shift(-1)=="Entry")|(Test["Short_Trade"].shift(-1) == "No trade")))]
            
            
                    trade_short.append(df1[['Time','stock1', 'Stock1_Open', 'Stock1_Close','stock2', 'Stock2_Open', 'Stock2_Close',"Short_Trade",
                                        "short Sell stock1","short Buy stock2","short stock1 qty","short stock2 qty","short Buy stock1","short Sell stock2","Correlation","Cointegrtion(pvalue)"]])
    
                    df2 = Test[((Test["Long_Trade"].shift(-1) == "Entry") & ((Test["Long_Trade"].shift(-2) == "Exit")|(Test["Long_Trade"].shift(-2) == "No trade")))|(((Test["Long_Trade"].shift(-2) == "Exit")) & ((Test["Long_Trade"].shift(-1)=="Entry")|(Test["Long_Trade"].shift(-1) == "No trade")))]
    
                    trade_long.append(df2[['Time','stock1', 'Stock1_Open', 'Stock1_Close','stock2', 'Stock2_Open', 'Stock2_Close',"Long_Trade",
                                       "long Buy stock1","long Sell stock2","long stock1 qty","long stock2 qty","long Sell stock1","long Buy stock2","Correlation","Cointegrtion(pvalue)"]])
    
                    Train_collect.append(Train)
                    Test_collect.append(Test)
    
```

```python
shorttrade=[]
for i in trade_short:
    i.loc[:,"Entry"]=np.where((i.loc[:,"short Sell stock1"] == 0),i.loc[:,"Time"].shift(-1),i.loc[:,"Time"])
    i.loc[:,"short Sell stock1"] = np.where((i.loc[:,"short Sell stock1"] == 0),i.loc[:,"short Sell stock1"].shift(-1),i.loc[:,"short Sell stock1"])
    i.loc[:,"short Buy stock2"] = np.where((i.loc[:,"short Buy stock2"] == 0),i.loc[:,"short Buy stock2"].shift(-1),i.loc[:,"short Buy stock2"])
    i.loc[:,"short stock1 qty"] = np.where((i.loc[:,"short stock1 qty"] == 0),i.loc[:,"short stock1 qty"].shift(-1),i.loc[:,"short stock1 qty"])
    i.loc[:,"short stock2 qty"] = np.where((i.loc[:,"short stock2 qty"] == 0),i.loc[:,"short stock2 qty"].shift(-1),i.loc[:,"short stock2 qty"])
    i.loc[:,"stock1 profit"] = i.loc[:,"short Sell stock1"] - i.loc[:,"short Buy stock1"]
    i.loc[:,"stock1 Tprofit"] = i.loc[:,"stock1 profit"] * i.loc[:,"short stock1 qty"]
    i.loc[:,"stock2 profit"] = i.loc[:,"short Sell stock2"] - i.loc[:,"short Buy stock2"]
    i.loc[:,"stock2 Tprofit"] = i.loc[:,"stock2 profit"] * i.loc[:,"short stock2 qty"]
    i["Investment"] = 200000
    i.loc[:,"Tprofit"] = round(i.loc[:,"stock1 Tprofit"] + i.loc[:,"stock2 Tprofit"],2)
    i.loc[:,"Tprofit%"]=round((i.loc[:,"Tprofit"]/200000)*100,2)
    i.loc[:,"Tdays"]=i.loc[:,"Time"]-i.loc[:,"Entry"]
    shorttrade.append(i[(i != 0).all(1)].dropna())
#trade_short[0][trade_short[0]["short Buy stock1"] != 0]
```
